```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Alphabets_data.csv')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X=data.drop('letter', axis=1)
y=data['letter']
```

```python
X_normalized=scaler.fit_transform(X)
```

```python
X_normalized=pd.DataFrame(X_normalized, columns=X.columns)
```

```python
X_normalized.head()
```

# EDA

```python
X_normalized.hist(bins=20,figsize=(20,15))
plt.suptitle("Distribution of Each Feature")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y)
plt.title('Class Distribution')
plt.xlabel('Alphabets')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=X_normalized, orient='h')
plt.title('Boxplot of Each Feature')
plt.xlabel('Feature Values')
plt.ylabel('Features')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_normalized.corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Outliers Tretment

```python
Q1=X_normalized.quantile(0.25)
Q3=X_normalized.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
X_filtered=X_normalized[~((X_normalized<lower_bound)|(X_normalized>upper_bound)).any(axis=1)]
```

```python
print(X_filtered.shape)
```

# One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
y_filtered = y.loc[X_filtered.index]
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
y_encoded = encoder.fit_transform(y_filtered.values.reshape(-1, 1))
```

```python
print(y_encoded.shape)
```

# Split the Dataset:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_filtered,y_encoded,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# ANN Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units=64, activation='relu',input_shape=(X_train.shape[1],)))
```

```python
model.add(Dense(units=32,activation='relu'))
```

```python
model.add(Dense(units=y_train.shape[1], activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

# Train the Model

```python
history=model.fit(X_train,y_train,epochs=50,batch_size=32,validation_split=0.2)
```

```python
test_loss, test_accuracy=model.evaluate(X_test,y_test)
```

```python
print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f}')
```

```python
import keras_tuner as kt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def build_model(hp):
  model=Sequential()
  model.add(Dense(units=hp.Int('units',min_value=32,max_value=512,step=32),activation='relu',input_shape=(X_train.shape[1],)))
  model.add(Dense(units=hp.Int('units', min_value=32, max_value=512, step=32), activation='relu'))
  model.add(Dense(units=y_train.shape[1], activation='softmax'))
  model.compile(optimizer=hp.Choice('optimizer', values=['adam', 'sgd']), loss='categorical_crossentropy', metrics=['accuracy'])
  return model
```

```python
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory='my_dir',
    project_name='helloworld'
  )
```

```python
tuner.search(X_train, y_train, epochs=50, validation_split=0.2)
```

```python
best_model=tuner.get_best_models(num_models=1)[0]
```

```python
test_loss,test_accuracy=best_model.evaluate(X_test,y_test)
```

```python
print(f'Test Loss: {test_loss:.4f}')
print(f'Test Accuracy: {test_accuracy:.4f}')
```

# Evaluation
